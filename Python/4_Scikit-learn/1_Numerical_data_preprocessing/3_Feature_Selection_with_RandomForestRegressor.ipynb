# Feature selection with random forests
Feature importance is measured as averaged information gain from all decision trees in a random forest.  
The algorithm works also with nonlinearly separable data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

## Load the dataset

```python
df = pd.read_excel('../Datasets/Boston.xls')
```

## Explore the dataset

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

## Define features and target variables

```python
colnames = df.columns.values
x = df[colnames[0:13]]
y = df[colnames[13]]
```

## Split into train and test sets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

## Train the model

```python
rfr = RandomForestRegressor(n_estimators=100)
rfr.fit(x_train,y_train)
```

## Evaluate the model

```python
rfr.score(x_test,y_test)
```

## Feature importance

```python
fi = pd.DataFrame({'Feature':colnames[0:13],'Importance': rfr.feature_importances_})
fi.sort_values('Importance',ascending=False)
```

## Correlation with target variable

```python
df.corr()['MEDV'].abs().sort_values(ascending=False)
```
