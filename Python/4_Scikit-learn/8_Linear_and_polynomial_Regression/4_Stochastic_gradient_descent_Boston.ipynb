```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.model_selection import train_test_split, validation_curve
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
```

## Load the dataset

```python
df = pd.read_excel('../Datasets/Boston.xls')
```

## Explore the dataset

```python
df.head()
```

```python
df.describe()
```

```python
df = df.sample(frac=1) # to shuffle the dataset
```

## Scatter Matrix

```python
sbn.pairplot(df,size=1, plot_kws={'s': 5})
plt.show()
```

## Correlation with target variable

```python
df.corr()['MEDV'].abs().sort_values(ascending=False)
```

## Select the features

```python
cols = df.columns.values
cols
```

```python
x = df[cols[0:13]]
y = df[cols[13]]
```

## Split into train and test sets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

## Scale the features

```python
#Standardize features by removing the mean and scaling to unit variance. 
#Always do it when using stochastic gradient descent.
x_scaler = StandardScaler()
x_train = x_scaler.fit_transform(x_train)
x_test = x_scaler.transform(x_test)
```

## Learning rate dependence

```python
estimator = SGDRegressor(learning_rate='constant')

param_range = [0.0001, 0.001, 0.01, 0.1, 1.0]

train_scores, test_scores = validation_curve(estimator, x_train, y_train, cv=10, param_name='eta0', 
                                             param_range= param_range ,n_jobs=-1)
```

```python
train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
```

```python
plt.plot(param_range, train_scores_mean, 'o-', color='r', label='Training score')
plt.plot(param_range, test_scores_mean, 'o-', color='g', label='Test score')
plt.xlabel('eta0')
plt.xscale('log')
plt.ylabel('Score')
#plt.ylim([0.1,1.01])
plt.legend(loc='best')
plt.show()
```

## Train the model

```python
model = SGDRegressor(learning_rate='constant',eta0=0.01)
model.fit(x_train, y_train)
```

## Evaluate the model

```python
model.score(x_test,y_test)
```

## Make predictions

```python
y_pred = model.predict(x_test)
```

## Compare data with prediction

```python
plt.scatter(y_test, (y_test - y_pred))
plt.plot([0,55],[0,0],c='r')
plt.xlim([0,55])
plt.xlabel('$y_{test}$',fontsize=20)
plt.ylabel('$y_{test}-y_{pred}$',fontsize=20)
plt.show()
```
