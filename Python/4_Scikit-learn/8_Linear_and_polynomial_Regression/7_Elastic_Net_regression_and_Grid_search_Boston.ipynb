# Elastic Net regression (L1+L2)

Linear regression with combined L1 and L2 regularization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import ElasticNet
from sklearn.preprocessing import StandardScaler
```

## Load the dataset

```python
df = pd.read_excel('../Datasets/Boston.xls')
```

## Explore the dataset

```python
df.head()
```

```python
df.describe()
```

```python
df = df.sample(frac=1) # to shuffle the dataset
```

## Scatter Matrix

```python
sbn.pairplot(df,size=1, plot_kws={'s': 5})
plt.show()
```

## Correlation with target variable

```python
df.corr()['MEDV'].abs().sort_values(ascending=False)
```

## Select the features

```python
cols = df.columns.values
cols
```

```python
x = df[cols[0:13]]
y = df[cols[13]]
```

## Split into train and test sets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

## Scale the features

```python
#Standardize features by removing the mean and scaling to unit variance. 
#Always do it when using stochastic gradient descent.
x_scaler = StandardScaler()
x_train = x_scaler.fit_transform(x_train)
x_test = x_scaler.transform(x_test)
```

## Hyperparameter optimization with cross-validation

```python
estimator = ElasticNet()

parameters = {'alpha': (0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000), 'l1_ratio':(0.1,0.2,0.4,0.5,0.6,0.7,0.8,0.9,1.0)}

grid_search = GridSearchCV(estimator, parameters, cv =5, scoring='r2', n_jobs=-1)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_score_
```

```python
 grid_search.best_params_
```

## Evaluate the best model

```python
model = grid_search.best_estimator_
#model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

## Make predictions using the best model

```python
y_pred = model.predict(x_test)
```

## Compare data with prediction

```python
plt.scatter(y_test, (y_test - y_pred))
plt.plot([0,55],[0,0],c='r')
plt.xlim([0,55])
plt.xlabel('$y_{test}$',fontsize=20)
plt.ylabel('$y_{test}-y_{pred}$',fontsize=20)
plt.show()
```
