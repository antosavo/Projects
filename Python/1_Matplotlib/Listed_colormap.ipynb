```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import colors
```

```python
data = np.loadtxt('Trajectories_Max/trajectories.dat')
x = data[:,0]
y = data[:,1]
z = data[:,2]
```

```python
#CMAP = LinearSegmentedColormap.from_list('mycmap', [(0, 'white'),(0.002, 'yellow'),(0.11, 'orange'),(0.90, 'red'),(1, 'black')])


# make a color map of fixed colors
CMAP = colors.ListedColormap(['white','yellow','orange','red','black'])
bounds=[0,1,5,9.9,101,105]
norm = colors.BoundaryNorm(bounds, CMAP.N)

plt.scatter(x,y,cmap=CMAP,c=z,marker='.',edgecolor='none',norm=norm)#vmin=0, vmax=105)
plt.xlabel('$\\it t$ [ps]',fontsize=16)#,labelpad=-4) #italic
plt.ylabel('$\\it z$ [m]',fontsize=16)#,labelpad=-1)
plt.xlim([15,50])
plt.ylim([200,450])
#plt.xlim([-25,75])
#plt.ylim([100,500])
#plt.xticks([6,7,8])
#plt.yticks([100,140,180,220])

plt.tick_params(labelsize=14,width=1)
cbar=plt.colorbar()
cbar.set_label('$\\it |u|^2$ [W]',fontsize=16)
cbar.set_ticks([0,10,100])
cbar.ax.tick_params(labelsize=14,width=1) 
plt.tight_layout()
plt.savefig('Trajectories_C.png')
plt.show()
```
